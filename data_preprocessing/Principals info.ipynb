```python
import json 

def clean_genres(x):
    json_dict = json.loads(x.replace("\'", "\""))
    return [item['name'] for item in json_dict][:num_genres]

def clean_prod_companies(x):
    companies = []
    i = x.find("'name': '")
    while (i != -1):
        companies.append(x[i + len("'name': '"):x.find("', 'origin_country'", i)])
        i = x.find("'name': '", i + 1)
        
    return companies[:num_prod_companies]
#     x_no_single_quotes = x.replace("\'id\'", '\"id\"').replace("\'logo_path\'", '\"logo_path\"').replace("\'origin_country\'", '\"origin_country\"').replace("\'name\'", '\"name\"').replace() 
#     y = x_no_single_quotes.replace(', \"origin_country\": \'\'', "").replace(', \"logo_path\": None', "")
#     print(y)
#     print(y[20:25])
#     json_dict = json.loads(y)
#     return [item['name'] for item in json_dict][:5]


movies_clean_genres = movies_dropped.copy()
movies_clean_genres['genres_x'] = movies_clean_genres['genres_x'].apply(clean_genres)

movie_clean_companies = movies_clean_genres.copy()
movie_clean_companies['production_companies'] = movie_clean_companies['production_companies'].apply(clean_prod_companies)
```

```python
import numpy as np
import math

mean_rating_by_year = dict.fromkeys(np.arange(firstYear, lastYear + 1), 0)
for i in range(firstYear, lastYear + 1):
    mean_rating_by_year[i] = movies_rev_budget_ratio[movies_rev_budget_ratio['startYear'] == i][['vote_average']].mean()[0]
    if math.isnan(mean_rating_by_year[i]) or mean_rating_by_year[i] == 0:
        mean_rating_by_year[i] = 5.0
```

```python
mean_rating_by_year
```

```python
def get_component_score(components, instance, score_type, cur_year):
    score = 0
    cur_year_int = cur_year[['startYear']].iloc[0][0]
    
    weight = len(components)
    for comp in components:
        relevant_cols = cur_year[cur_year[comp] == instance][score_type]
        if score_type == 'vote_average':
            for v in list(relevant_cols.values):
                score += (v - mean_rating_by_year[cur_year_int]) * weight
        else:
            score += sum(list(relevant_cols.values)) * weight
        weight -= 1
    
    return score
```

```python
%%time
import numpy
import statistics

total_scores_keys = []
for col_type in nontrivial_col_types:
    for metric in success_metrics:
        total_scores_keys.append(col_type[0][:-2] + '|' + metric)
total_scores = dict.fromkeys(total_scores_keys, 0)

individual_scores = {}

movies_encoded = movies.copy()
for metric in success_metrics:
    for col_type in nontrivial_col_types:
        for col in col_type:
            movies_encoded[col + '_' + metric] = 0.0
        
mean_std_last_year = {}
        
for i in range(firstYear, lastYear + 1):
    print(i)
    cur_year = movies_encoded[movies_encoded['startYear'] == i]

    for col_type in nontrivial_col_types:
        for metric in success_metrics:            
            all_ids = set()
            rows = cur_year[col_type].values
            for row in rows:
                for id_ in row:
                    if id_ != '' and id_ not in all_ids:
                        all_ids.add(id_)

            for id_ in all_ids:
                key_code = col_type[0][:-2] + '|' + metric + '|' + id_
                if key_code not in individual_scores:
                    individual_scores.update({key_code: 0})
                score = get_component_score(col_type, id_, metric, cur_year)
                individual_scores[key_code] +=  score
                total_scores[col_type[0][:-2] + '|' + metric] += score
                
    normalized_scores = individual_scores.copy()            
    
    for col_type in nontrivial_col_types:
        for metric in success_metrics:
            all_values = []
            
            for key in individual_scores.keys():
                if key.find(col_type[0][:-2] + '|' + metric) != -1:
                    all_values.append(individual_scores[key])
                    
            if len(all_values) < 2:
#                 print(i)
                for key in normalized_scores.keys():
                    normalized_scores[key] = 0
                continue
                    
            std = statistics.stdev(all_values)
            mean = statistics.mean(all_values)
            
            if i == lastYear:
                mean_std_last_year[col_type[0][:-2] + '|' + metric] = mean, std
                
            for key in normalized_scores.keys():
                if key.find(col_type[0][:-2] + '|' + metric) != -1:
                    normalized = individual_scores[key]
                    normalized -= mean
                    normalized /= std
                    if(std != 0):
                        normalized_scores[key] = normalized
                    else:
                        normalized_scores[key] = 0
    for index, row in cur_year.iterrows():
        for col_type in nontrivial_col_types:
            for metric in success_metrics:
                for col in col_type:
                    id_ = row[col]
                    if id_ != '':            
                        key_code = col_type[0][:-2] + '|' + metric + '|' + id_
                        movies_encoded.set_value(index, col + '_' + metric, normalized_scores[key_code])            
```
