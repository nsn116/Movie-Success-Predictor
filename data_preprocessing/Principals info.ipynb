```python
movies_dropped = imdb_tmdb.drop(['belongs_to_collection', 'Unnamed: 0', 'id', 'imdb_id',
               'original_language', 'overview', 'popularity', 'production_countries',
               'release_date', 'spoken_languages', 'status', 'status_code', 'status_message',
               'tagline', 'original_title', 'titleType', 'primaryTitle', 'isAdult',
               'release_year', 'runtime', 'genres_y', '_merge'], axis=1)
```

```python
num_genres = 4
num_prod_companies = 5
num_directors = 3
num_writes = 5
num_actors = 10
```

```python
import json 

def clean_genres(x):
    json_dict = json.loads(x.replace("\'", "\""))
    return [item['name'] for item in json_dict][:num_genres]

def clean_prod_companies(x):
    companies = []
    i = x.find("'name': '")
    while (i != -1):
        companies.append(x[i + len("'name': '"):x.find("', 'origin_country'", i)])
        i = x.find("'name': '", i + 1)
        
    return companies[:num_prod_companies]
#     x_no_single_quotes = x.replace("\'id\'", '\"id\"').replace("\'logo_path\'", '\"logo_path\"').replace("\'origin_country\'", '\"origin_country\"').replace("\'name\'", '\"name\"').replace() 
#     y = x_no_single_quotes.replace(', \"origin_country\": \'\'', "").replace(', \"logo_path\": None', "")
#     print(y)
#     print(y[20:25])
#     json_dict = json.loads(y)
#     return [item['name'] for item in json_dict][:5]


movies_clean_genres = movies_dropped.copy()
movies_clean_genres['genres_x'] = movies_clean_genres['genres_x'].apply(clean_genres)

movie_clean_companies = movies_clean_genres.copy()
movie_clean_companies['production_companies'] = movie_clean_companies['production_companies'].apply(clean_prod_companies)
```

```python
movies_joined_crew = pd.merge(movie_clean_companies, crew, left_on='tconst', right_on='tconst')
```

```python
%%time
principals_name_basics = principals_no_job_chars.copy()

cnt = 0 

new_rows = []

for index, row in name_basics.iterrows():
    cnt += 1
    if cnt % 10000 == 0:
        print(cnt)
#         break

    if row['nconst'] not in relevant_principals:
        continue

    if not isinstance(row['knownForTitles'], str):
        continue
    
    
    titles = row['knownForTitles'].split(',')
    primaryProfession = row['primaryProfession'].split(',')
    isActor = primaryProfession[0] == 'actor' or primaryProfession[0] == 'actress'
    
    for t in titles:
        new_rows.append(pd.Series([t, 11, row['nconst'], primaryProfession[0]], index=principals_name_basics.columns))
```

```python
len(new_rows)
```

```python
%%time
principals_name_basics = principals_name_basics.append(new_rows, ignore_index=True)
```
