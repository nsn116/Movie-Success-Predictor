```python
import pandas as pd
```

```python
from glob import glob

jsons = []

for f_name in glob('data/*.json'):
    jsons.append(pd.read_json(f_name))
```

```python
df_unioned = pd.concat(jsons)
df_1 = df_unioned.drop_duplicates(subset=['imdb_id'])
```

```python
df_2 = pd.read_csv('./title_cleaned.csv')
```

```python
genres = df_1[['genres']]
s = set()
for i in range(genres.values.shape[0]):
    if isinstance(genres.values[i][0], list):
        for j in range(len(genres.values[i][0])):
            l = []
            l.append(genres.values[i][0][j]['name'])
            s = s.union(set(l))
print(s)
```

```python
title_filter = lambda x : x.upper().replace('...', '')

df_1_released = df_1[df_1['status'] == 'Released']
df_1_released['release_year'] = df_1_released['release_date'].apply(lambda x : x[:4])
df_1_released['title'] = df_1_released['title'].apply(title_filter)
```

```python
# should be redundant
df_2_movies = df_2[(df_2['titleType'] == 'movie') | (df_2['titleType'] == 'tvMovie')]
```

```python
df_2_movies.count()
```

```python
df_2_released_movies = df_2_movies[df_2_movies['startYear'] != '\\N']
```

```python
df_2_released_movies.count()
```

```python
df_2_released_movies['startYear'] = df_2_released_movies['startYear'].apply(lambda x : str(x))
df_2_released_movies['primaryTitle'] = df_2_released_movies['primaryTitle'].apply(title_filter)
```

```python
def extract_country(x):
    try:
        return x[0]['name']
    except IndexError:
        return 'Invalid'
    
df_1_countries = df_1_released.copy()
df_1_countries['production_countries'] = df_1_countries['production_countries'].apply(extract_country)
df_1_us = df_1_countries[df_1_countries['production_countries'] == 'United States of America']
```

```python
joined_inner = pd.merge(df_1_us, df_2_released_movies,
                  left_on=['imdb_id'], right_on=['tconst'],
                  how='inner', indicator=True)
```

```python
joined_inner_non_zero = joined_inner[(joined_inner['budget'] != 0) & (joined_inner['revenue'] != 0)]
```

```python
joined_inner_non_zero.to_csv('../data/imdb_tmdb.csv')
```
