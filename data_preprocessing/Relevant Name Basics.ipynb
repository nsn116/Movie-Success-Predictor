```python
import pandas as pd
```

```python
imdb_tmdb = pd.read_csv('../data/imdb_tmdb.csv')
```

```python
relevant_titles = set(imdb_tmdb['tconst'].values)
```

```python
name_basics = pd.read_csv('name_basics.tsv', delimiter='\t')
```

```python
def filter_relevant_titles(x):
    known_for_titles = set(x.split(','))
    intersection = known_for_titles.intersection(relevant_titles)
    return known_for_titles.intersection(relevant_titles)
```

```python
name_basics['knownForTitles'] = name_basics['knownForTitles'].apply(filter_relevant_titles)
```

```python
name_basics
```

```python
relevant_name_basics = name_basics[name_basics['knownForTitles'] != set()]
```

```python
def convert_to_str(x):
    return str(x)[1:-1].replace("'", '')
```

```python
relevant_name_basics['knownForTitles'] = relevant_name_basics['knownForTitles'].apply(convert_to_str)
```

```python
relevant_name_basics.to_csv('../data/relevant_name_basics.csv')
```
